## TLDR;

```python
from zerosum_env import make
from random import choice
from zerosum_env.envs.carbon.helpers import *

def my_agent(observation, configuration):

  board = Board(observation, configuration)
  me = board.current_player

  # 
  for worker in me.collectors:
    worker.next_action = WorkerAction.UP
  
  # 
  for worker in me.planters:
    worker.next_action = WorkerAction.DOWN
  
  # 
  for worker in me.collectors:
    worker.next_action = WorkerAction.LEFT
  
  # 
  for worker in me.planters:
    worker.next_action = WorkerAction.RIGHT

  # 
  for recrtCenter in me.recrtCenters:
    recrtCenter.next_action = RecrtCenterAction.RECCOLLECTOR

  # 
  for recrtCenter in me.recrtCenters:
    recrtCenter.next_action = RecrtCenterAction.RECPLANTER
      
  # 
  return me.next_actions

def agent(obs):
    action = {}
    worker_id = list(obs.players[obs.player][1].keys())[0]
    if obs.players[obs.player][2] != {}:
        worker_id = list(obs.players[obs.player][2].keys())[0]
#     'RECCOLLECTOR',
    worker_action = choice([ 'RECPLANTER', 'UP', 'DOWN', 'RIGHT', 'LEFT', None])
    if worker_action is not None:
        action[worker_id] = worker_action
#     print(action)
    return action

env = make("carbon", debug=True)
env.run([agent, my_agent])
env.render(mode="ipython", width=1000, height=800)
# print(env.render(mode="json"))
```
